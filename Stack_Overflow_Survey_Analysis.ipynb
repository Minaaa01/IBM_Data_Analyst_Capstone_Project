```python
# Function to fill NaN values with random non-NaN values from the same column
def fill_nan_with_random_values(df):
    for col in df.columns:
        nan_indices = df[df[col].isnull()].index  # Get indices of NaN values in the column
        non_nan_values = df[col].dropna().values  # Get non-NaN values from the column
        random_values = np.random.choice(non_nan_values, size=len(nan_indices))  # Choose random non-NaN values
        df.loc[nan_indices, col] = random_values  # Fill NaN values with random non-NaN values
```

## Question 18

**What is the Most Languages Worked with**

```python
# QUERY="""
# Select sql
# FROM sqlite_schema
# WHERE name = 'LanguageWorkedWith'
# """
# dff = pd.read_sql_query(QUERY,conn) # Table that i will work on
# print(dff.iat[0,0]) #Get value at specified row/column pair
QUERY = """
SELECT LanguageWorkedWith, count(LanguageWorkedWith) as Count
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
"""
dff = pd.read_sql_query(QUERY,conn)
dff.sort_values('Count', ascending=False,inplace=True)
dff = dff.head(10) # get top five
dff.plot.bar(x='LanguageWorkedWith')
plt.xlabel('LanguageWorkedWith')
plt.ylabel('Respondent')
plt.show()
```
